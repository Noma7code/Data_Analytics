## The data
### **netflix_data.csv**
| Column | Description |
|--------|-------------|
| `show_id` | The ID of the show |
| `type` | Type of show |
| `title` | Title of the show |
| `director` | Director of the show |
| `cast` | Cast of the show |
| `country` | Country of origin |
| `date_added` | Date added to Netflix |
| `release_year` | Year of Netflix release |
| `duration` | Duration of the show in minutes |
| `description` | Description of the show |
| `genre` | Show genre `[|]()`

```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
print(netflix_df.head(6))

ninty=netflix_df[(netflix_df["release_year"]>=1990) & (netflix_df["release_year"]<2000)]#filter movies released in the 1990's
duration1=ninty["duration"] #subset the duration of the 1990's movies
plt.hist(duration1) #  histogram to visualize the data to find the most frequent duration
plt.title("Distribution of 1990s movie duration")
plt.xlabel("Duration(minutes)")
plt.ylabel("No. of Movies")
plt.show()
duration=100  #The most frequent movie duration is 100 min, from the histogram

count=0
action=ninty[ninty["genre"]=="Action"]#subset the data for action movies
for label,row in action.iterrows():
    if row["duration"]<90:
       count=count+1
short_movie_count=count    
print(short_movie_count)#The number of short actions movies from the 1990s

# The genre of movies that is mostly watched and those that are not likely to be watched
top_mov_genre=netflix_df['genre'].value_counts()
print(top_mov_genre)

#The data of top 5 country of origin of the movies dominated on the netflix platform
top_mov_country=netflix_df['country'].value_counts()
print(top_mov_country.head())

#subset the titles of the movie
print(netflix_df['title'].head(10))

# movie description
best_description=netflix_df['description']
print(best_description.head(5))
    
```
