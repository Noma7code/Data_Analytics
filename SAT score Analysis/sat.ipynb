Every year, American high school students take SATs, which are standardized tests intended to measure literacy, numeracy, and writing skills. There are three sections - reading, math, and writing, each with a **maximum score of 800 points**. These tests are extremely important for students and colleges, as they play a pivotal role in the admissions process.

Analyzing the performance of schools is important for a variety of stakeholders, including policy and education professionals, researchers, government, and even parents considering which school their children should attend. 

You have been provided with a dataset called `schools.csv`, which is previewed below.

You have been tasked with answering three key questions about New York City (NYC) public school SAT performance.

```python
# import the pandas package with the alias pd 
import pandas as pd

# Read in the data
schools = pd.read_csv("schools.csv")

# Preview the data
print(schools.head(5))

#subset the school name and its average math score
best_math_schools1 = schools[["school_name", "average_math"]]

# best in maths must score 80% of 800, that is 640
best_math_schools = best_math_schools1[best_math_schools1["average_math"] >= 640].sort_values("average_math", ascending=False) 

#school with the best math score
print(best_math_schools)

#the total SAT Score for each school
schools["total_SAT"] = schools["average_math"] + schools["average_writing"] + schools["average_reading"]
print(schools.head())

#top 10 performing school based on total SAT score
top_10_schools = schools.sort_values("total_SAT", ascending=False)[["school_name", "total_SAT"]].head(10)
print(top_10_schools.head(10))

#bourough with the largest standard deviation
borough_by = schools.groupby("borough")["total_SAT"].agg(["count", "mean", "std"]).round(2)
print(borough_by)
largest_std_dev = borough_by[borough_by["std"] == borough_by["std"].max()]
largest_std_dev = largest_std_dev.rename(columns={"count": "num_schools", "mean": "average_SAT", "std": "std_SAT"})
print(largest_std_dev)

```
